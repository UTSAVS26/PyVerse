# Building the ML model

Logistic Regression, Decision Trees, Random Forest, and Gradient Boosting are all supervised machine learning algorithms used for different types of tasks, such as classification and regression. Let's briefly explain each of these algorithms:

1. **Logistic Regression:**
   - **Type:** Classification
   - **Use Case:** Used for binary or multiclass classification problems.
   - **How it works:** Despite its name, logistic regression is a classification algorithm. It models the probability that an instance belongs to a particular category. It's especially useful for binary classification tasks (e.g., spam or not spam).

2. **Decision Trees:**
   - **Type:** Classification and Regression
   - **Use Case:** Used for both classification and regression problems.
   - **How it works:** Decision Trees split the dataset into subsets based on the most significant attribute at each step. Each node in the tree represents a decision based on an attribute, and each leaf node represents the outcome.

3. **Random Forest:**
   - **Type:** Ensemble method (Bagging)
   - **Use Case:** Used for classification and regression tasks.
   - **How it works:** Random Forest is an ensemble of multiple decision trees. Each tree is trained on a random subset of the data, and the final prediction is the result of combining the predictions of all individual trees. This helps to reduce overfitting and improve accuracy.

4. **Gradient Boosting:**
   - **Type:** Ensemble method (Boosting)
   - **Use Case:** Used for classification and regression tasks.
   - **How it works:** Gradient Boosting builds an ensemble of decision trees sequentially. Each tree corrects the errors of the previous one. It focuses on instances that were misclassified by the previous trees, gradually improving the model's performance.

In summary, Logistic Regression is a simple linear model for classification, Decision Trees are versatile for both classification and regression, Random Forest is an ensemble of decision trees for improved accuracy, and Gradient Boosting is an ensemble method that builds trees sequentially to correct errors and enhance predictive performance. The choice of which algorithm to use depends on the specific characteristics of the data and the problem at hand.

## The datasets are then split into a training and testing set in a 70:30 ratio.

We’d use the confusion matrix, accuracy score, f1 score, and cross-validation score to evaluate performance.
**Confusion matrix**:- used to describe the performance of a classification model.
- **True Positive (TP)**:- people with diabetes that are correctly predicted
- **False Positive (FP)**:- non-diabetics that are incorrectly predicted as diabetic
- **True Negative (TN)**:- non-diabetics that are correctly predicted
- **False Negative (FN)**:- people with diabetes that are incorrectly predicted as non-diabetic

**Accuracy**:- fraction of correct predictions by the model — (TP +TN) / (TP + TN + FP +FN)

**F1 score**:- harmonic average of precision and recall — 2 x ((Precision x Recall) / (Precision + Recall))

**Cross-validation score**:- trains ML models on a subset of the dataset and evaluates another subset.

```python
# Split the dataset into 70% Training set and 30% Testing set
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
def classification_model_performance(y_test, prediction, model, x_train, y_train):
    accuracy = metrics.accuracy_score(y_test, prediction)
    print("Accuracy Score: {:.3%}".format(accuracy))
    f1 = metrics.f1_score(y_test, prediction)
    print("F1 Score: {:.3%}".format(f1))
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    accuracy_model = []
    for train_index, test_index in kf.split(x_train):
        X_train, X_test = x_train.iloc[train_index], x_train.iloc[test_index]
        Y_train, Y_test = y_train.iloc[train_index], y_train.iloc[test_index]
        model.fit(X_train, Y_train)
        accuracy_model.append(accuracy_score(Y_test, model.predict(X_test)))
    print("Cross-Validation Score: {:.3%}".format(np.mean(accuracy_model)))

```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def confusion_matrix_plot(y_test, prediction):
    cm = confusion_matrix(y_test, prediction)
    classes = ['0', '1']
    fig, ax = plt.subplots(figsize=(8, 6))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
    disp.plot(cmap=plt.cm.Blues, values_format='d', ax=ax)
    plt.title('Confusion Matrix', fontsize=16)
    plt.xlabel('Predicted Label', fontsize=14)
    plt.ylabel('True Label', fontsize=14)
    cbar = disp.im_.colorbar
    cbar.set_label('Count', rotation=270, labelpad=15, fontsize=12)
    
    # Adjust tick labels and font size
    ax.set_xticklabels(classes, fontsize=12)
    ax.set_yticklabels(classes, fontsize=12)
    plt.show()
```

## Logistic Regression Model

```python
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
prediction = logmodel.predict(x_test)
classification_model_performance(y_test, prediction, logmodel, x_train, y_train)

confusion_matrix_plot(y_test, prediction)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, f1_score, make_scorer
from sklearn.model_selection import StratifiedKFold

def evaluate_models(models, X_train, y_train, X_test, y_test):
    """
    Evaluate multiple machine learning models and plot their performance metrics.

    Parameters:
    - models (dict): Dictionary containing model names as keys and corresponding model instances as values.
    - X_train, y_train, X_test, y_test: Training and testing data.

    Returns:
    - None (plots the performance metrics).
    """

    metrics = ['Accuracy', 'F1 Score', 'Cross-Validation']

    # Initialize lists to store performance metrics for each model
    accuracy_scores = []
    f1_scores = []
    cv_scores = []

    # Define a stratified k-fold cross-validation strategy
    cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    # Evaluate each model
    for model_name, model in models.items():
        # Train the model
        model.fit(X_train, y_train)

        # Make predictions
        y_pred = model.predict(X_test)

        # Calculate performance metrics
        accuracy = accuracy_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        cv = np.mean(cross_val_score(model, X_train, y_train, cv=cv_strategy, scoring=make_scorer(accuracy_score)))

        # Append metrics to lists
        accuracy_scores.append(accuracy)
        f1_scores.append(f1)
        cv_scores.append(cv)

    # Plot the performance metrics with annotations
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

    for i, metric in enumerate(metrics):
        bars = axes[i].bar(models.keys(), [accuracy_scores, f1_scores, cv_scores][i], color=['skyblue', 'lightcoral', 'lightgreen'])
        axes[i].set_ylabel(metric)
        axes[i].set_title(f'{metric} Comparison')

        # Annotate the bars with their values
        for bar in bars:
            yval = bar.get_height()
            axes[i].text(bar.get_x() + bar.get_width()/2, yval, round(yval, 3), ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

```

## Plotting the Metrics score of every model

```python

models_to_evaluate = {'Logistic Regression': logmodel, 'Decision Tree': dec_tree, 'Random Forest': rfmodel, 'Gradient Boosting': gb}
evaluate_models(models_to_evaluate, x_train, y_train, x_test, y_test)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36", "Upgrade-Insecure-Requests":"1"}
URL = "https://www.flipkart.com/oppo-find-n3-flip-sleek-black-256-gb/product-reviews/itm25e330f3e4f41?pid=MOBGTR89PFVGSHS7&lid=LSTMOBGTR89PFVGSHS7EEISBD&marketplace=FLIPKART" 
page = requests.get(URL, headers=headers)
request1 = BeautifulSoup(page.content, "html.parser")
request1
```

```python
# fetch product name

name = request1.find('div', {'class':'Vu3-9u eCtPz5'}).get_text()
print(name)
```

```python
reviews = request1.find_all('div',{'class':'ZmyHeo'})
```

```python
reviews
```

```python
for review in reviews:
    print("* " + review.get_text())
    print('------------------')
```

```python
# fetch average rating

ratings = request1.find('div', {'class':'ipqd2A'}).get_text()
print("Average Rating:- " + ratings)
```

```python
# fetch all ratings

individual_ratings = request1.find_all('div',{'class':'XQDdHH Ga3i8K'})
for rating in individual_ratings:
    print(" - " + rating.get_text())
```

```python
# fetch customer names

customers = request1.find_all('p',{'class':'_2NsDsF AwS1CA'})
print('Customer Names: \n')
for customer in customers:
    print(customer.get_text())
```

```python
# fetch product price

price = request1.find('div',{'class':'Nx9bqj'})
print('Product Price: '+price.get_text())
```

```python
# fetch comments

comments = request1.find_all('p',{'class':'z9E0IG'})
for comment in comments:
    print(' - ' + comment.get_text())
```
